```python
import pandas as pd
from autogluon.tabular import TabularPredictor

# Load Model
predictor = TabularPredictor.load(
    "AutoGluonModels/ag-20260609_084918"
)

# Load Dataset
df = pd.read_csv("dataset/cargo_breakdown_dataset_30000.csv")

# Take one sample row
sample = df.drop(
    columns=["Breakdown_Time_Minutes"]
).head(1)

print("Input Data:")
print(sample)

prediction = predictor.predict(sample)

print("\nPredicted Breakdown Time:")
print(prediction)
```
